# Nordic-Baltic Team Ratings & League Rankings

Runs the `build_nordic_baltic_ratings.py` pipeline from the `hradeck-scouting` repo on Wyscout data for **Sweden, Denmark, Norway, Finland, Iceland, Latvia, and Lithuania**.

Produces:
1. **Team Attacking / Defensive / Overall ratings** (0-100) for every club, percentiled within its own league, plus a tier-adjusted score for cross-league comparison.
2. **League Rankings** — how the 15 leagues/divisions in the region compare to each other.

Run the cells top to bottom. If the repo is private you'll be prompted for a GitHub personal access token on first run.

```python
import os, subprocess

REPO = "marclamberts/hradeck-scouting"
BRANCH = "claude/nordic-baltic-league-data-769b60"  # update once merged to main

if not os.path.isdir("hradeck-scouting"):
    ret = subprocess.run(["git", "clone", "-b", BRANCH, f"https://github.com/{REPO}.git"])
    if ret.returncode != 0:
        from getpass import getpass
        token = getpass("Repo looks private - paste a GitHub personal access token: ")
        subprocess.run(["git", "clone", "-b", BRANCH, f"https://{token}@github.com/{REPO}.git"], check=True)

%cd hradeck-scouting
```
